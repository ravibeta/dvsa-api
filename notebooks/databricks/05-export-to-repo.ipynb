# 05 · Export to Databricks Repos & CI/CD (dbx)

How to add these notebooks to Databricks Repos and deploy them with `dbx`.

## Add to Databricks Repos
1. Repos > Add Repo > `https://github.com/ravibeta/DVSA-APIs`.
2. Notebooks live under `notebooks/databricks/`.

## Deploy with dbx (commented — run from your laptop, not the notebook)
```bash
pip install dbx
dbx configure --profile DEFAULT
# Deploy the batch job defined in jobs/job_template.json:
dbx deploy --deployment-file jobs/job_template.json
dbx launch --job dvsa-batch-inference
```
